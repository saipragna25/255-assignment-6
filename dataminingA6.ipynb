```python
import opendatasets as od
import pandas as pd
import numpy as np
from mlxtend.frequent_patterns import association_rules, apriori
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import fpgrowth
```

```python

dataset=pd.read_csv('bread basket.csv')
dataset.head()
```

```python
dataset.shape
```

```python
dataset.Item.nunique(), dataset.Transaction.nunique()
```

```python
dataset_transactions = dataset.groupby(['Transaction', 'Item'])['Item'].count().reset_index(name='Frequency')
dataset_transactions
```

```python
transactions = dataset_transactions.pivot_table(index='Transaction', columns='Item', values='Frequency', aggfunc='sum').fillna(0)
transactions.head()
```

```python
def binary_encode(x):
    if x == 0:
        return 0
    if x >= 1:
        return 1

transactions = transactions.applymap(binary_encode)
transactions.head()
```

```python
frequent_sets = apriori(transactions, min_support=0.005, use_colnames=True)
```

```python
# Rules from frequent itemset 
rules = association_rules(frequent_sets, metric='lift', min_threshold=1)
# Sort values by confidence
rules.sort_values('confidence', ascending=False, inplace=True)
rules
```
